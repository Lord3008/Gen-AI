```python
import requests
import io
from PIL import Image

# Hugging Face API details
API_URL = "https://api-inference.huggingface.co/models/CompVis/stable-diffusion-v1-4"
headers = {"Authorization": "Bearer <---Enter your own API key---->"}  # Replace with your actual token

# Function to send the query to Hugging Face model
def query(payload):
    try:
        print(f"Sending payload: {payload}")  # Debugging: Print the payload
        response = requests.post(API_URL, headers=headers, json=payload)
        print(f"Response Status Code: {response.status_code}")  # Print status code
        
        if response.status_code == 200:
            return response.content  # Successful response
        else:
            print(f"Error: {response.text}")  # Print error message if not successful
            return None
            
    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred: {err}")  # Print the error
        return None
    except Exception as e:
        print(f"An error occurred: {e}")  # Print any other exceptions
        return None

# Example input
image_bytes = query({
    "inputs": "A cat on a sea beach",  # Test input
})

# If the query was successful, open and display the image
if image_bytes:
    try:
        image = Image.open(io.BytesIO(image_bytes))
        image.show()  # Display the image
    except Exception as e:
        print(f"Error processing the image: {e}")  # Print any image processing errors
else:
    print("Failed to generate an image.")

```

```python
import requests
import io
from PIL import Image
import gradio as gr
import warnings

warnings.filterwarnings("ignore")

# Hugging Face API details
API_URL = "https://api-inference.huggingface.co/models/CompVis/stable-diffusion-v1-4"
headers = {"Authorization": "Bearer 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"}  # Replace with your actual token

# Function to send the query to Hugging Face model
def query(payload):
    try:
        response = requests.post(API_URL, headers=headers, json=payload)
        if response.status_code == 200:
            return response.content  # Successful response
        else:
            return f"Error: {response.text}"  # Print error message if not successful
    except Exception as e:
        return f"An error occurred: {e}"  # Print any other exceptions

# Function to generate an image from a text prompt
def generate_image(prompt):
    image_bytes = query({"inputs": prompt})  # Use user input prompt
    if isinstance(image_bytes, bytes):  # Check if the response is bytes (successful)
        return Image.open(io.BytesIO(image_bytes))  # Convert bytes to an image
    else:
        return None  # Return None if there was an error

# Create a Gradio interface
iface = gr.Interface(
    fn=generate_image,  # Function to be called
    inputs=gr.Textbox(label="Enter your prompt:", placeholder="e.g., A beautiful landscape with mountains and a river"),
    outputs=gr.Image(label="Generated Image"),
    title="Text-to-Image Generator",
    description="Generate images based on textual descriptions using Hugging Face models.",
    allow_flagging="never"
)

# Launch the Gradio app
iface.launch()
```
